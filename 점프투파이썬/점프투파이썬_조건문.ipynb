# if 문의 기본 구조
```
if 조건문:
    수행할_문장1
    수행할_문장2
    ...
else:
    수행할_문장A
    수행할_문장B
    ...
```

# 조건문이란 ?
- 참과 거짓을 판단하는 문장

## 비교 연산자
| 비교연산자 |	설명 |
|:----------:|:-----:|
|x < y	|x가 y보다 작다.|
|x > y	|x가 y보다 크다.|
|x == y	|x와 y가 같다.|
|x != y	|x와 y가 같지 않다.|
|x >= y	|x가 y보다 크거나 같다.|
|x <= y	|x가 y보다 작거나 같다.|

## and, or, not
| 연산자 |	설명 |
|:------:|:-----:|
|x or y	|x와 y 둘 중 하나만 참이어도 참이다.|
|x and y	|x와 y 모두 참이어야 참이다.|
|not x	|x가 거짓이면 참이다.|

## in, not in
- 영어 단어 in의 뜻이 ‘~안에’라는 것

| in |	not in |
|:----:|:-----:|
|x in 리스트|	x not in 리스트|
|x in 튜플|	x not in 튜플|
|x in 문자열|	x not in 문자열|

## 비교 연산자의 연쇄 사용
- 비교 연산자의 연쇄 사용

## 조건부 표현식의 기본 형태
```
변수 = 참일_때_값 if 조건 else 거짓일_때_값
```

````python
pocket = ['paper', 'cellphone']

if 'money' in pocket :
  print('택시를 타라')
elif 'card' in pocket :
  print('택시를 타라')
else :
  print('걸어가라')
````

````python
# if 문 한 줄로 작성하기
pocket = ['paper', 'money', 'cellphone']

if 'money' in pocket : pass
else : print('카드를 꺼내라')
````

````python
# 비교 연산자의 연쇄 사용

x = 5
print(1 < x < 10)
print((1<x) and (x<10))
print(10 <= x <= 20)
print((10<=x) and (x<=20))
````

````python
# 조건부 표현식

score = 85

if score >= 60 :
  result = '합격'
else :
  result = '불합격'

print(result)

# 한 줄로 표현하기
score = 85
result = '합격' if score >= 60 else '불합격'

print(result)
````
