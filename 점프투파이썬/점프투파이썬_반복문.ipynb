# while 문의 기본 구조
```
while 조건문:
    수행할_문장1
    수행할_문장2
    수행할_문장3
    ...

```

## while 문 강제로 빠져나가기
- 강제로 멈추게 하는 것이 바로 `break 문`

## while 문의 맨 처음으로 돌아가기
- while 문을 빠져나가지 않고 while 문의 맨 처음(조건문)으로 다시 돌아가게 만들고 싶은 경우가 생기게 된다. 이때 사용하는 것이 바로 `continue 문`

## 무한루프
- 무한히 반복한다는 의미

### 무한루프의 기본 형태
```
while True:
    수행할_문장1
    수행할_문장2
    ...

```

## while-else 문
- while 문이 정상적으로 종료되었을 때(break로 빠져나가지 않았을 때) else 절이 실행된다.
- break 문으로 while 문을 빠져나가면 else 절은 실행되지 않는다.

````python
treeHit = 0
while treeHit < 10 :
  treeHit += 1
  print("나무를 %d번 찍었습니다." % treeHit)
  if treeHit == 10:
    print('나무가 넘어갑니다.')
````

위 예에서 while 문의 조건문은 `treeHit < 10`이다.   
즉, treeHit가 10보다 작은 동안 while 문에 포함된 문장들을 계속 수행

````python
prompt = """
1. Add
2. Del
3. List
4. Quie
"""
number = 0

while number != 4 :
  print(prompt)
  number = int(input())
  print(f'{number}번을 선택하였습니다.')
````

````python
coffee = 10
money = 300

while money :
  print('돈을 받았으니 커피를 줍니다.')
  coffee -= 1
  print(f'남은 커피의 양은 {coffee}개 입니다.')
  if coffee == 0:
    print('커피가 다 떨어졌습니다. 판매를 중지합니다.')
    break
````

````python
coffee = 3

while True :
  money = int(input('돈을 넣어주세요. : '))
  if money == 300 :
    print('커피를 줍니다.')
    coffee -= 1
  elif money > 300 :
    print(f'거스름돈 {money-300}원을 주고 커피를 줍니다.')
    coffee = coffee -1
  else :
    print('돈을 다시 돌려주고 커피를 주지 않습니다.')
    print(f'남은 커피의 양은 {coffee}개 입니다.')
  if coffee == 0:
    print('커피가 다 떨어졌습니다. 판매를 중지합니다.')
    break
````

````python
a = 0

while a < 10 :
  a += 1
  if a % 2 == 0 : continue
  print(a)
````

````python
count = 0

while count < 3 :
  print(f'카운트 : {count}')
  count += 1
else :
  print('while 문이 정상 종료되었습니다.')
````

````python
count = 0

while count < 5 :
  if count == 2 :
    break
  print(f'카운트 : {count}')
  count += 1
else : # break 문으로 while 문을 빠져나가면 else 절은 실행되지 않는다.
  print('while 문이 정상 종료되었습니다.')
````

## 중첩된 while 문
- while 문 안에 또 다른 while 문을 사용할 수 있다. 이를 중첩된 while 문

````python
i = 1
while i <= 3:
  j = 1
  while j <= 3 :
    print(f'i={i}, j={j}')
    j += 1
  i += 1
````

## for 문과 함께 자주 사용하는 range 함수
- for 문은 숫자 리스트를 자동으로 만들어 주는 range 함수

````python
# 전형적이 for문
test_list = ['one', 'two', 'three']
for i in test_list :
  print(i)
````

````python
# 다양한 for문 사용

a = [(1,2), (3,4), (5,6)]
for (first, last) in a :
  print(first + last)
````

````python
# for문 응용

marks = [90, 25, 67, 45, 80]
number = 0

for i in marks :
  number = number +1
  if i >= 60 :
    print(f'{number}번 학생 합격')
  else :
    print(f'{number}번 학생 불합격')
````

````python
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks :
  number += 1
  if mark < 60 :
    continue
  print(f'{number}번 학생 축하합니다. 합격입니다. *^^*')
````

````python
a = range(10) #  0부터 10 미만의 숫자를 포함
print(a)
````

````python
add = 0

for i in range(1,11) :
  add += i

print(add)
````

````python
marks = [90, 25, 67, 45, 80]

for number in range(len(marks)) :
  if marks[number] < 60 :
    continue
  print("%d번 학생 축하합니다. 합격입니다." % (number+1))
````

````python
for i in range(2,10) :
  for j in range(1,10) :
    print(f'{i} * {j} = {i*j}', end= " ")
  print('')
````

````python
for i in range(2, 10):         # 2단부터 9단까지
  print(f'[{i}단]')            # 단 제목 출력
  for j in range(1, 10):       # 1부터 9까지 곱하기
    print(f'{i} * {j} = {i*j}')
  print('')
````

# 리스트 컴프리헨션 사용하기
- 리스트 안에 for 문을 포함하는 리스트 컴프리헨션(list comprehension)을 사용하면 좀 더 편리하고 직관적인 프로그램을 만들 수 있다.
- `[표현식 for 항목 in 반복_가능_객체 if 조건문]`
- for 문을 2개 이상 사용하는 것도 가능
  - ```
    [표현식 for 항목1 in 반복_가능_객체1 if 조건문1
      for 항목2 in 반복_가능_객체2 if 조건문2
      ...
      for 항목n in 반복_가능_객체n if 조건문n]
    ```

````python
a = [1,2,3,4]
result = []

for num in a :
  result.append(num*3)

print(result)
````

````python
a = [1,2,3,4]

result = [num*3 for num in a]
print(result)
````

````python
a = [1,2,3,4]
result = [num*3 for num in a if num % 2 == 0]
print(result)
````

````python
# 컴프리헨션으로 구구단 만들기

result = [i*j for i in range(2,10)
for j in range(1,10)]

print(result)
````

## for 문과 break 문

- while 문에서와 마찬가지로 for 문에서도 break 문을 사용
- break 문은 for 문을 강제로 빠져나가고 싶을 때 사용

````python
for i in range(10) :
  if i == 5 :
    break
  print(i)
````

## for-else 문
- else 절을 사용할 수 있다.
- for 문이 정상적으로 종료되었을 때(break로 빠져나가지 않았을 때) else 절이 실행된다.

````python
for i in range(5) :
  print(i)
else :
  print('for 문이 정상 종료 되었습니다.')
````

````python
for i in range(5):
  if i == 3 :
    break
  print(i)
else :
  print('for 문이 정상 종료 되었습니다.')
````

## enumerate 함수 활용하기
- 리스트의 순서(인덱스)와 값을 함께 구하고 싶을 때는 `enumerate 함수`를 사용하면 편리
- enumerate는 0부터 시작하는 인덱스 번호를 자동으로 생성

````python
fruits = ['apple', 'banana', 'orange']

for i, fruit in enumerate(fruits) :
  print(f'{i} : {fruit}')
````

````python
fruits = ['apple', 'banana', 'orange']

for i, fruit in enumerate(fruits, 1) : # 시작 번호를 변경하고 싶다면 이렇게
  print(f"{i}: {fruit}")
````

## zip 함수로 여러 리스트 함께 순회하기
- 두 개 이상의 리스트를 동시에 순회하고 싶을 때는 zip 함수를 사용한다.

````python
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for name, score in zip(names, scores) :
  print(f'{name}님 : {score}점')
````

````python
names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]

for name, kor,eng in zip(names, korean, english) :
  print(f'{name} 학생 : 국어 {kor}점, 영어 {eng}점')
````
