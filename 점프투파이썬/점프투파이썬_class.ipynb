```python
# 더하기 구현

result = 0

def add(num) :
  global result
  result += num
  return result

print(add(3))
print(add(4))
```

```python
result1 = 0
result2 = 0

def add1(num) :
  global result1
  result1 += num
  return result1

def add2(num) :
  global result2
  result2 += num
  return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

```python
# 클래스로 구현하기

class Calculator :
  def __init__(self) :
    self.result = 0

  def add(self, num) :
    self.result += num
    return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

```python
# 덧셈 기능

class FourCal() :
  def setdata(self, first, second) :
    self.first = first
    self.second = second

  def add(self) :
    result = self.first + self.second
    return result

a = FourCal()
a.setdata(4,2)
print(a.add())
```

```python
# 곱하기, 빼기, 나누기 기능

class FourCal() :
  def setdata(self, first, second) :
    self.first = first
    self.second = second

  def add(self) :
    result = self.first + self.second
    return result

  def mul(self) :
    result = self.first * self.second
    return result

  def sub(self) :
    result = self.first - self.second
    return result

  def div(self) :
    result = self.first / self.second
    return result
```

```python
a = FourCal()
b = FourCal()

a.setdata(4,2)
b.setdata(3,8)
```

```python
# 생성자 사용하기

class FourCal() :
  def __init__(self, first, second) :
    self.first = first
    self.second = second

  def add(self) :
    result = self.first + self.second
    return result

  def mul(self) :
    result = self.first * self.second
    return result

  def sub(self) :
    result = self.first - self.second
    return result

  def div(self) :
    result = self.first / self.second
    return result
```

```python
c = FourCal(4,2)
c.first
```

```python
c.second
```

```python
c.add()
```

```python
# 상속

class MoreFourCal(FourCal) :
  pass

a = MoreFourCal(4,2)
print(a.add())
```

```python
class MoreFourCal(FourCal) :
  def pow(self) :
    result = self.first ** self.second
    return result

a = MoreFourCal(4,2)
print(a.pow())
```
