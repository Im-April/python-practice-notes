# 사용자 입출력
좀 더 다양한 입출력 방법에 대해서 알아보자.

## 사용자 입력 활용하기
## 1. input() 사용하기

```python
# input() 사용하기
a = input()
print(a)
```

```python
# 프롬프트를 띄워 사용자 입력받기

a = input('간단한 인사를 입력하세요 : ')
print(a)
```

```python
# input은 입력되는 모든 것을 문자열로 취급

num = input('숫자를 입력하세요 : ')
print(type(num))
print(num)
```

```python
# 입력값을 숫자로 변환하기
age = input("나이를 입력하세요: ")
age = int(age) # 문자열을 정수로 변환

print(age+1)
```

```python
# 실수로 변환하기

height = input("키를 입력하세요(cm): ")
height = float(height) # 문자열을 실수로 변환
print(height/100) # 미터 단위로 변환
```

```python
# 한 번에 변환하기

age = int(input("나이를 입력하세요: "))
print(age)
print(type(age))
```

# print 자세히 알기
- 지금까지 우리가 사용한 print 문의 용도는 데이터를 출력하는 것

## 1. 큰따옴표로 둘러싸인 문자열은 + 연산과 동일하다

## 2. 문자열 띄어쓰기는 쉼표로 한다

## 3. sep 매개변수로 구분자 설정하기

## 4. 한 줄에 결괏값 출력하기

```python
# 1. 큰따옴표로 둘러싸인 문자열은 + 연산과 동일하다

print("life" "is" "too short")  # 1번
print('-'*30)
print("life"+"is"+"too short")  # 2번
```

```python
# 문자열 띄어쓰기는 쉼표로 한다

print("life", "is", "too short") # 쉼표(,)를 사용하면 문자열을 띄어 쓸 수 있다.
```

```python
# sep 매개변수로 구분자 설정하기

print("2025", "08", "17", sep="-")
print("점프", "투", "파이썬", sep=" TO ")
```

```python
# 한 줄에 결괏값 출력하기

for i in range(10) :
  print(i,end=' ')
```

```python
# calculator.py
print("=== 간단한 계산기 ===")

# 사용자로부터 두 숫자 입력받기
num1 = float(input("첫 번째 숫자를 입력하세요: "))
num2 = float(input("두 번째 숫자를 입력하세요: "))

# 계산 결과 출력
print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")

if num2 != 0:
    print(f"{num1} / {num2} = {num1 / num2}")
else:
    print("0으로 나눌 수 없습니다.")
```
